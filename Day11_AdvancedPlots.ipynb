```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('final_merged.csv')
print(df.head())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='BP', y='Cholesterol', hue='Gender', style='Smoker')
plt.title('BP vs Cholesterol by Gender & Smoker Status')
plt.xlabel('Blood Pressure')
plt.ylabel('Cholesterol')
plt.savefig('Day11_BPvsCholesterol_Gender.png')
plt.show()
```

```python
grouped = df.groupby(['Gender', 'Smoker']).agg({'BP':'mean'}).reset_index()

plt.figure(figsize=(8,6))
sns.barplot(data=grouped, x='Gender', y='BP', hue='Smoker')
plt.title('Average BP by Gender & Smoker')
plt.ylabel('Average BP')
plt.savefig('Day11_AvgBP_Barplot.png')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='Smoker', y='Cholesterol', palette='Set2')
plt.title('Cholesterol Distribution by Smoker Status')
plt.savefig('Day11_Boxplot_Cholesterol.png')
plt.show()
```
